```python
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv.imread('/content/tommm.jpeg')
print(type(img))
print(img.shape)
```

```python
cv2_imshow(img)
```

```python
grayimage=cv.cvtColor(img,cv.COLOR_BGR2GRAY )
cv2_imshow(grayimage)
```

```python
inverted_image = 255 - grayimage
cv2_imshow(inverted_image)
```

```python
blurred = cv.GaussianBlur(inverted_image,(25,25),0)
cv2_imshow(blurred)
```

```python
inverted = 255 -blurred
pencilsketch=cv.divide(grayimage,inverted ,scale=256.0)
cv2_imshow(pencilsketch)
```
